# Scatter plot

##### Read data from file

```python
data=open('Salary_Data.csv','r').read()
lines=data.split('\n')
lines
x=[]
y=[]
for line in lines:
    if len(line)>1:
        l=line.split(',')
        x.append(float(l[0]))
        y.append(float(l[1]))

```

```python
plt.plot(x,y,'go')
plt.xlabel('Salary')
plt.ylabel('Years of experience')
plt.show()
```

```python
plt.bar(x,y)
```

# Pie plot

```python
days=[1,2,4]
slices=[7,2,15]
eating=[2,3,4,5]
sleeping=[7,8,7,6]
working=[14,11,10,9]
plt.pie(days,labels=['eating','sleeping','working'],startangle=90,shadow=True,explode=(1,0,0),autopct='%1.1f%%')
plt.show()
```

# Live chart

```python
import matplotlib.animation as animation
import numpy as np
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
data=np.loadtxt('new.txt',unpack=True,delimiter=',')
x=data[0]
y=data[1]
ax1.clear()
ax1.plot(x,y)
ani=animation.FuncAnimation(fig,ax1,interval=1000)
plt.show()
```
